```python
def solution(K : int,A : list[int]) -> int :
    result = 0
    cur = 0
    for i in range(len(A)):
        if A[i] >= K:
            result += 1
        elif A[i] + cur >= K:
            result += 1
            cur = 0
        elif A[i] + cur < K:
            cur += A[i]
    return result
```

```python
# 테스트 케이스 1: 기본적인 예제
K = 4
A = [1, 2, 3, 4, 1, 1, 3]
print(solution(K,A))
# 예상 출력: 3

# 테스트 케이스 2: 이미 모든 로프의 길이가 K보다 클 때
K = 3
A = [4, 5, 6, 7, 8]
# 모든 로프의 길이가 이미 K보다 크므로 묶을 필요가 없음
# 예상 출력: 5
print(solution(K,A))

# 테스트 케이스 3: 모든 로프의 길이가 K보다 작을 때
K = 10
A = [1, 2, 3, 4, 5]
# 모든 로프를 묶어야만 K 이상의 로프가 될 수 있음
# 예상 출력: 1
print(solution(K,A))

# 테스트 케이스 4: 로프 길이가 1일 때
K = 1
A = [1, 1, 1, 1]
# 모든 로프가 이미 K 이상이므로 묶지 않아도 됨
# 예상 출력: 4
print(solution(K,A))

# 테스트 케이스 5: 로프의 길이가 모두 동일하고 묶지 않아도 되는 경우
K = 3
A = [3, 3, 3, 3]
# 모든 로프가 K와 같으므로 각 로프는 개별로 유지
# 예상 출력: 4
print(solution(K,A))

# 테스트 케이스 6: 배열 길이가 1일 때
K = 2
A = [3]
# 하나의 로프만 있고 K보다 크므로
# 예상 출력: 1
print(solution(K,A))

# 테스트 케이스 7: 로프를 최대한 묶을 필요가 있는 경우
K = 7
A = [5, 2, 6, 1, 5, 1]
# 5 + 2 = 7, 6 + 1 = 7, 5 + 1 = 6 -> 총 2개의 로프
# 예상 출력: 2
print(solution(K,A))

# 테스트 케이스 8: 로프를 최대한 묶어도 K 이상이 안 되는 경우
K = 100
A = [1, 1, 1, 1, 1, 1, 1, 1]
# 로프를 아무리 묶어도 100 이상이 안 됨
# 예상 출력: 0
print(solution(K,A))

# 테스트 케이스 9: 중간에 아주 큰 값이 있는 경우
K = 10
A = [1, 1, 50, 1, 1, 1]
# 50은 이미 K 이상, 나머지 1들을 묶어도 10이 안 됨
# 예상 출력: 1
print(solution(K,A))

# 테스트 케이스 10: K가 1인 경우
K = 1
A = [1, 1, 1, 1, 1]
# 모든 로프가 이미 K 이상이므로 묶을 필요 없음
# 예상 출력: 5
print(solution(K,A))

# 테스트 케이스 11: 빈 배열인 경우는 없으므로 N=1인 경우로 대체
K = 1000000000
A = [1000000000]
# 배열에 하나의 로프가 있고 그 값이 K와 같으므로
# 예상 출력: 1
print(solution(K,A))


#35144120151
```
